```python
#Parameters
prev = 0.01 #Prevalence
sens = 0.95
spec = 0.90
```

```python
#The prior is prevalence

#Posterior
def posterior(prevalence: Union[float, np.array] = 0.01, sensitivity: Union[float, np.array] = 0.95, specificity: Union[float, np.array] = 0.90) -> Union[float, np.array]:
    """
    Description: Calculates the posterior probability of actually having disease X given a positive test result for parameters: prevalence, sensitivity, and specificity.
    
    Parameters:
    -----------
        prevalence: float or np.array (optional) - Is the prevalence of disease X.
        sensitivity: float or np.array (optional) - Is the sensitivity of the test for disease X.
        specificity: float or np.array (optional) - Is the specificity of the test for disease X.
        
    Returns:
    --------
        float or np.array - The posterior probabilityh of actually having disease X given a positive test result.
    """
    
    return (sensitivity * prevalence) / (sensitivity * prevalence + (1 - specificity) * (1 - prevalence))
```

```python
def multivariate_normal_density(x, mu, Sigma):
    """
    Returns the density of a D-dimensional vector(s) x given a D-dimensional mean vector mu
    and a DxD-dimensional covariance matrix Sigma, without using scipy.stats.multivariate_normal.pdf().

    Args:
        x: A numpy array of shape (N, D) representing N D-dimensional vectors, or (D,) for a single vector.
        mu: A numpy array of shape (D,) representing the D-dimensional mean vector.
        Sigma: A numpy array of shape (D, D) representing the DxD covariance matrix.

    Returns:
        A numpy array of shape (N,) containing the density of each vector in x, or a single float if x is a single vector.
        Returns a numpy array of zeros if Sigma is not positive definite.
        Raises a ValueError if the input dimensions are incorrect.
    """

    x = np.asarray(x)
    mu = np.asarray(mu)
    Sigma = np.asarray(Sigma)

    D = mu.shape[0]

    if x.ndim == 1:
        if x.shape != (D,):
            raise ValueError("x must be a 1D array of length D when a single vector is given.")
        x = x.reshape(1, D) # Reshape to (1,D) to handle single vector as a row vector
        N = 1
    elif x.ndim == 2:
        if x.shape[1] != D:
            raise ValueError("x must be a 2D array where the second dimension is D when multiple vectors are given.")
        N = x.shape[0]
    else:
        raise ValueError("x must be a 1D or 2D array.")

    if mu.shape != (D,):
        raise ValueError("mu must be a 1D array of length D")
    if Sigma.shape != (D, D):
        raise ValueError("Sigma must be a 2D array of shape (D, D)")

    try:
        Sigma_inv = np.linalg.inv(Sigma)
        det_Sigma = np.linalg.det(Sigma)
    except np.linalg.LinAlgError:
        return np.zeros(N)

    density = np.zeros(N)

    for i in range(N):
        x_minus_mu = x[i] - mu
        exponent = -0.5 * np.dot(np.dot(x_minus_mu, Sigma_inv), x_minus_mu)
        normalization = 1 / np.sqrt((2 * np.pi) ** D * det_Sigma)
        density[i] = normalization * np.exp(exponent)

    if N == 1:
        return density[0]
    else:
        return density
```

```python
#Spherical Gaussian
d = 2
mu = np.zeros(shape=(d,))
cov = np.identity(d)
x,y = np.mgrid[-1:1:100j, -1:1:100j]
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)

z = np.column_stack([x.flatten(), y.flatten()])

llm_pdf = multivariate_normal_density(z, mu, cov)
print("Shape of Gemini's PDF:", llm_pdf.shape)

sp_pdf = multivariate_normal.pdf(z, mean=mu, cov=cov)
print("Shape of Scipy's PDF:", sp_pdf.shape)

llm_pdf = llm_pdf.reshape(x.shape)
sp_pdf = sp_pdf.reshape(x.shape)

print(f"Mean of percent difference of Gemini's PDF with Scipy's PDF: {np.mean((np.abs(llm_pdf - sp_pdf) / sp_pdf) * 100)}")
```

```python
#3D Plot of Gemini's Spherical Gaussian Distribution
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, llm_pdf, cmap="cool")
ax.set_title("Gemini's Spherical Gaussian Distribution")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("PDF")
plt.show()
```

```python
#3D Plot of Scipy's Spherical Distribution
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(x, y, sp_pdf, cmap="cool")
ax.set_title("Scipy's Spherical Gaussian Distribution")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("PDF")
plt.show()
```
