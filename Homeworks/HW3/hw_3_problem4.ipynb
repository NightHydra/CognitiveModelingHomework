```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import stan

try:
    import arviz as az
except ImportError as err:
    print("Please, install arviz for easy visualization of Stan models.")

import nest_asyncio
nest_asyncio.apply()

import pandas as pd
```

```python
mu0 = 5
sigma0 = 1

sigma2 = 3

number_of_datapoints_to_simulate = 1000
```

```python
prior = np.random.normal(mu0, sigma0, size=number_of_datapoints_to_simulate)

posterior = prior + np.random.normal(0, sigma2, size=number_of_datapoints_to_simulate)
```

```python
sns.histplot({"Prior": prior, "Posterior": posterior}, kde=True, stat="probability")
```
