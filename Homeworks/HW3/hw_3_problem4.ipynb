```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import stan

try:
    import arviz as az
except ImportError as err:
    print("Please, install arviz for easy visualization of Stan models.")

import nest_asyncio
nest_asyncio.apply()

import pandas as pd
```

```python
mu0 = 200
sigma0 = 500

sigma2 = 100
```

```python
def generate_priors(mu0 : float = 0, sigma0 : float = 1, num_priors : int = 100):

    return np.random.normal(mu0, sigma0, num_priors)

def sample_posterior_given_prior(prior_samples : np.array, sigma2 : float = 1) -> np.array:

    return np.random.normal(prior_samples, sigma2)
```

```python
def plot_prior_and_posterior(prior_arr : np.array, posterior_arr):
    f1, ax1 = plt.subplots(1, 1, figsize=(8, 4))
    sns.histplot(prior_arr, ax = ax1, color = "cyan", alpha = 0.8, label = "Priors")

    f2, ax2 = plt.subplots(1, 1, figsize=(8, 4))
    sns.histplot(posterior_arr, ax = ax2, color = "maroon", alpha = 0.8, label = "Posteriors")
```

```python
prior_samples = generate_priors(mu0, sigma0, 1000)

posterior_samples = sample_posterior_given_prior(prior_samples, sigma2)

plot_prior_and_posterior(prior_samples, posterior_samples)
```
